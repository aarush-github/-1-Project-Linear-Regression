# Prediction using Supervised ML
Task: Predict the percentage of an student based on the number of study hours.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
datalink = "http://bit.ly/w-data"
data = pd.read_csv(datalink)
```

```python
data.head()
```

```python
X = data['Hours'].values.reshape(-1,1)
y = data['Scores'].values.reshape(-1,1)
```

```python
plt.plot(X,y,'o')
plt.xlabel('hours')
plt.ylabel('Scores')
plt.grid()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state =0, test_size = 0.3)
```

```python
clf = LinearRegression()
linearReg = clf.fit(X_train, y_train) 
```

```python
m = linearReg.coef_
c = linearReg.intercept_
a = m*X + c
```

```python
plt.plot(X,y, 'o')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.plot(X, a, '-')
plt.grid()
plt.show()
```

```python
y_pred = linearReg.predict(X_test)
```

```python
dict = {'Actual': y_test.flatten(), 'Prediction': y_pred.flatten()}
df = pd.DataFrame(dict)
df
```

```python
print('Linear Regression score = ', clf.score(X_test, y_test))
print('Mean Squared Error = ', metrics.mean_squared_error(y_test, y_pred))
print('Mean Absolute Error = ', metrics.mean_absolute_error(y_test, y_pred))
```
